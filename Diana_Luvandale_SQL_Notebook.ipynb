Creation of a Pandas Datafarme

```
import pandas as pd
import csv
# Loading our grandelectors dataset
granddata = pd.read_csv('/content/GrandElectors_by_state.csv')
# loading our population dataset
popdata = pd.read_csv('/content/Population_by_state.csv')
```

```
# Saving grandelectors dataframe into a database
%sql DROP TABLE if EXISTS popdata;
%sql PERSIST popdata;

%sql SELECT * FROM popdata;
```

```
# changes the to uppercase permanently
%%sql
UPDATE 
  granddata
SET
  state = UPPER(state)
```

Merging the datasets

```
# joining the tables
%%sql
CREATE VIEW merged AS
SELECT replace(popdata.State, 'DISTRICT OF COLUMBIA', 'D.C') as State, Population,granddata.GrandElectors
FROM popdata INNER JOIN granddata
    ON popdata.state = granddata.state
```

```
# Display the merged table
%%sql
SELECT * from merged;
```

Ratio calculation

```
# Adding column called ratio
%%sql
CREATE VIEW 
Ratios
AS
SELECT State, Population, GrandElectors, (GrandElectors *1.0/Population * 1.0) as Ratio 
from merged
ORDER BY Ratio DESC;
```

Finding the running sum of the grand electors

```
# Calculating running sum for the grand electors
%%sql 
CREATE VIEW 
RUNNING_SUM
AS
SELECT t1.*, sum(t2.GrandELectors) as RunningSum
from Ratios t1
INNER JOIN Ratios t2 on t1.Ratio <= t2.Ratio
GROUP BY t1.State
ORDER BY t1.Ratio ASC;
```

```
# Display the table with the running sum
%%sql
SELECT * from RUNNING_SUM;
```

```
# Filter out the states that have a running sum of 269 +1 
%%sql 
SELECT RUNNING_SUM.State from RUNNING_SUM 
WHERE RunningSum <= 271
ORDER BY RUNNING_SUM.Ratio DESC;
```
